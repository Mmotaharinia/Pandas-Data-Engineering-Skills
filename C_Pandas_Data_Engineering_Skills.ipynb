```python
#Merging dataframes- the how(outer,inner,left,right) argument basically determines what parts we wanna show based on a Venn diagram
staff_df=pd.DataFrame([{'name':'sally','role':'data manager'},
                         {'name':'alex','role':'director'},
                         {'name':'dave','role':'analyst'}])
staff_df=staff_df.set_index('name')

staff_df
```

```python
student_df=pd.DataFrame([{'name':'jen','department':'physics'},
                         {'name':'alex','department':'science'},
                         {'name':'sally','department':'chemistry'}])
student_df=student_df.set_index('name')
student_df
```

```python
#merge
pd.merge(staff_df,student_df,how='outer',right_index=True,left_index=True)
```

```python
pd.merge(staff_df,student_df,how='inner',right_index=True,left_index=True)
```

```python
pd.merge(staff_df,student_df,how='left',right_index=True,left_index=True)
```

```python
#a better way for merging using one column
pd.merge(staff_df,student_df, how='outer',on='name')
```

```python
#merge data with overlaping names but different family names
staff_df=pd.DataFrame([{'name':'sally','lastname':'choi','role':'data manager'},
                         {'name':'alex','lastname':'morales','role':'director'},
                         ])
staff_df=staff_df.set_index('name')
student_df=pd.DataFrame([{'name':'alex','lastname':'morales','department':'science'},
                         {'name':'sally','lastname':'morgan','department':'chemistry'}])
student_df=student_df.set_index('name')

pd.merge(staff_df,student_df, how='outer',on=['name','lastname'])
```
